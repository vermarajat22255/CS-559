```python
data = pd.read_csv("P4-Demographic-Data.csv")
print(data.columns)
# Distribution plotting 
sns.distplot(data["Internet users"], bins=50 )
```

```python
sns.boxplot(data=data, y="Birth rate", x="Income Group")
```

```python
#lmplot is linear model plot
sns.lmplot(data=data, x="Birth rate", y="Internet users", fit_reg=False, hue="Income Group",\
           size=5, aspect=2, markers="o")
```

---

#Seaborn is made on matplotlib library and internally calls its functions to create visualisations
#We dont have params in sns.lmplot() to increase the size of the marker 

                    #But
#sns provides a (Keywords) KWS dictionary to take all the additional parameters to matplotlib which have increase marker param
